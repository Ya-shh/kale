```python
n_rows = 100
output_path = "/tmp/dataset.csv"
```

```python
import csv
import random

with open(output_path, "w", newline="") as f:
    w = csv.writer(f)
    w.writerow(["x", "y"])
    for _ in range(n_rows):
        x = random.random()
        w.writerow([x, 2 * x + 0.1 * random.random()])
print(f"wrote {n_rows} rows to {output_path}")
```
