```python
dataset_path = "/tmp/dataset.csv"
model_path = "/tmp/model.txt"
epochs = 10
learning_rate = 0.001
```

```python
import csv

rows = list(csv.reader(open(dataset_path)))[1:]
n = len(rows)
slope = sum(float(r[1]) for r in rows) / max(n, 1)
with open(model_path, "w") as f:
    f.write(f"slope={slope:.4f}\nepochs={epochs}\nlr={learning_rate}\n")
print(f"trained on {n} rows for {epochs} epochs, slope={slope:.4f}")
```
